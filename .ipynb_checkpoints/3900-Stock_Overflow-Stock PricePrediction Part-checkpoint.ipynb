```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
plt.style.use("fivethirtyeight")
%matplotlib inline

from datetime import datetime
```

# 1.Read Tesla Data

```python
file_path = "Tesla.csv - Tesla.csv.csv"
tesla = pd.read_csv(file_path)
```

```python
tesla['Date'] = pd.to_datetime(tesla['Date'])
tesla.set_index("Date", inplace=True)
tesla.describe()
```

# 2. Basic Variables from this dataset (Not necessary for the prediction)

```python
#Calculate the moving average
ma_day = [10, 20, 50]

for ma in ma_day:
        column_name = f"MA for {ma} days"
        tesla[column_name] = tesla['Adj Close'].rolling(ma).mean()
```

```python
tesla[['Adj Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot()
plt.title(f"Tesla")
```

```python
#Daily return of the stock in average
tesla['Daily Return'] = tesla['Adj Close'].pct_change()
tesla['Daily Return'].plot(legend=True, linestyle='--', marker='o')
plt.title('Tesla')
```

# 3.Training the dataset

```python
data = tesla.filter(['Close'])
dataset = data.values

training_data_len = int(np.ceil( len(dataset) * .95 ))
training_data_len
```

```python
#Scale Data
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(dataset)

scaled_data
```

```python
train_data = scaled_data[0:int(training_data_len), :]

x_train = []
y_train = []

for i in range(60, len(train_data)):
    x_train.append(train_data[i-60:i, 0])
    y_train.append(train_data[i, 0])
    if i<= 61:
        print(x_train)
        print(y_train)
        print()
        
x_train, y_train = np.array(x_train), np.array(y_train)

x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
```

```python
# Build the LSTM model
model = Sequential()
model.add(LSTM(128, return_sequences=True, input_shape= (x_train.shape[1], 1)))
model.add(LSTM(64, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(x_train, y_train, batch_size=1, epochs=1)
```

# 4.Prediction

```python
test_data = scaled_data[training_data_len - 60: , :]

x_test = []
y_test = dataset[training_data_len:, :]
for i in range(60, len(test_data)):
    x_test.append(test_data[i-60:i, 0])
    
x_test = np.array(x_test)

x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1 ))

# Get the predicted price values 
predictions = model.predict(x_test)
predictions = scaler.inverse_transform(predictions)

# root mean squared error
rmse = np.sqrt(np.mean(((predictions - y_test) ** 2)))
rmse
```

```python
train = data[:training_data_len]
valid = data[training_data_len:]
valid['Predictions'] = predictions
# Visualization
plt.title('Model')
plt.xlabel('Date', fontsize=18)
plt.ylabel('Close Price USD ($)', fontsize=18)
plt.plot(train['Close'])
plt.plot(valid[['Close', 'Predictions']])
plt.legend(['Train', 'Val', 'Predictions'], loc='lower right')
plt.show()
```
